```python
import lib


n_values = [10, 1000, 100000, 1000000, 10000000, 50000000, 100000000, 500000000]

one_thread_time = []
multi_thread_time = []

print("1Thread")
for n in n_values:
    one_thread_time.append(lib.measure_time_spent(lambda: lib.integral_pi(n)))
    print(f"n: {n}, time spent: {one_thread_time[-1]}")

print("\n--------------------------\n")
print(f"{lib.THREAD_COUNT}Threads")
for n in n_values:
    multi_thread_time.append(lib.measure_time_spent(lambda: lib.integral_pi_parallel(n)))
    print(f"n: {n}, time spent: {multi_thread_time[-1]}")

print("\n--------------------------\n")
print("Speed increasing")
for i in range(len(n_values)):
    print(f"For n: {n_values[i]} in {one_thread_time[i] / multi_thread_time[i]}")
```
